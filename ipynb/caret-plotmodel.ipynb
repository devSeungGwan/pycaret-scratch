```python
from pycaret.datasets import get_data
from pycaret.classification import setup, create_model, plot_model
```

```python
diabetes = get_data('diabetes')
```

```python
clf = setup(data=diabetes, target='Class variable')
```

```python
lr = create_model('lr', round=10)
```

```python
plot_model(lr, plot="confusion_matrix")
```

```python
plot_model(lr, plot='threshold')
```

```python
plot_model(lr, plot="learning")
```
